```python
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
import pytesseract
```

```python
img = cv2.imread("id1-cropped.png")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
bfilter = cv2.bilateralFilter(gray, 5, 3, 3)
```

```python
plt.imshow(cv2.cvtColor(bfilter, cv2.COLOR_BGR2RGB))
```

```python
print(pytesseract.image_to_string(img))
```

```python

edged = cv2.Canny(bfilter, 30, 200)
```

```python
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
```

```python
print(pytesseract.image_to_string(edged))
```

```python
median_blur = cv2.medianBlur(bfilter, 5)
```

```python
plt.imshow(cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB))
```

```python
print(pytesseract.image_to_string(median_blur))
```

```python
gaussian_blur = cv2.GaussianBlur(bfilter, (5,5), 0)
```

```python
plt.imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB))
```

```python
print(pytesseract.image_to_string(gaussian_blur))
```

```python
median_blur = cv2.medianBlur(median_blur, 1)
```

```python
sharp_img = cv2.addWeighted(gray, 1.5, median_blur, -0.5, 3)
```

```python
plt.imshow(cv2.cvtColor(sharp_img, cv2.COLOR_BGR2RGB))
```

```python
print(pytesseract.image_to_string(sharp_img))
```

```python
canny_img = cv2.Canny(gray, 100,150)
```

```python
plt.imshow(cv2.cvtColor(canny_img, cv2.COLOR_BGR2RGB))
```

```python
print(pytesseract.image_to_string(canny_img))
```

```python
min_val = 80
max_val = 150

def change_max_val(val1):
    global max_val
    max_val = val1
    
def change_min_val(val2):
    global min_val
    min_val = val2
    
    
cv2.namedWindow("new")
cv2.createTrackbar('min_val',"new",0,500,change_min_val)
cv2.createTrackbar('max_val',"new",100,500,change_max_val)

while(True):
    canny_img = cv2.Canny(gray, min_val, max_val)

    cv2.imshow("new1",canny_img)
    print(pytesseract.image_to_string(canny_img))
    temp = cv2.waitKey(1)
    if(temp == ord('q')):
        break
        
cv2.destroyAllWindows()
```
